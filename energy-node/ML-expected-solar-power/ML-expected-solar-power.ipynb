```python
import kagglehub

DATASET = "anikannal/solar-power-generation-data"

dataset_path = kagglehub.dataset_download(DATASET)
print(f"Dataset downloaded to: {dataset_path}")
```

```python
import pandas as pd

raw_df_power = pd.read_csv(f"{dataset_path}/Plant_1_Generation_Data.csv")
raw_df_weather = pd.read_csv(f"{dataset_path}/Plant_1_Weather_Sensor_Data.csv")

raw_df_power
```

```python
raw_df_weather
```

```python
# convert the 'DATE_TIME' column to datetime format
raw_df_power['DATE_TIME'] = pd.to_datetime(raw_df_power['DATE_TIME'])
raw_df_weather['DATE_TIME'] = pd.to_datetime(raw_df_weather['DATE_TIME'])

# merge the two dataframes on 'DATE_TIME'
merged_raw_df = raw_df_power.merge(raw_df_weather, right_on='DATE_TIME', left_on='DATE_TIME', how='inner')
merged_raw_df
```

```python
# SELECT: DATE_TIME, AMBIENT_TEMPERATURE, MODULE_TEMPERATURE, IRRADIATION, AC_POWER
merged_df = merged_raw_df[['AMBIENT_TEMPERATURE', 'MODULE_TEMPERATURE', 'IRRADIATION', 'AC_POWER']]
merged_df
```

```python
merged_df.info()
```

```python
merged_df.describe()
```

```python
# train the model: regression with target AC_POWER

import numpy as np

# Prepare the data
target = np.array(merged_df.pop('AC_POWER'))
features = np.array(merged_df)

target
```

```python
features
```

```python
from sklearn.model_selection import train_test_split

train_features, test_features, train_target, test_target = train_test_split(features, target, test_size=0.2, random_state=42, shuffle=True)
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout

def rmse(y_true, y_pred):
    return tf.sqrt(tf.reduce_mean(tf.square(y_pred - y_true)))

# parameters
NB_neurones = 30
NB_features = train_features.shape[1]  # number of features in the dataset
activation_func = 'relu'
DROPOUT_rate = 0.4

# Define the model
model = Sequential([
    Dense(NB_neurones, activation=activation_func, name="Input", input_shape=(NB_features,),),
    Dense(NB_neurones, activation=activation_func, name="Dense_1"),
    Dense(NB_neurones, activation=activation_func, name="Dense_2"),
    Dense(NB_neurones, activation=activation_func, name="Dense_3"),
    # Dense(NB_neurones, activation=activation_func, name="Dense_4"),
    Dropout(DROPOUT_rate, name="Dropout"),
    Dense(1, name="Output"),
])

model.name = "Solar_Power_Prediction_Model"

# Compile the model
model.compile(optimizer='adam', loss=rmse, metrics=['mae'])

model.summary()
```

```python
# train the model
model.fit(train_features, train_target, epochs=40, validation_split=0.2, verbose=1, shuffle=True)
```

```python
# evaluate the model
performance = model.evaluate(test_features, test_target, verbose=1, steps=None)
print(f"Final RMSE: {performance[0]}, final MAE: {performance[1]}")
```
